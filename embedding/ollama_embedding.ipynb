```python
from langchain_community.embeddings import OllamaEmbeddings

embeds = OllamaEmbeddings(model='mxbai-embed-large')
```

```python
# query embedding

query = 'Whats second letter in Greek?'

vecs = embeds.embed_query(query)

len(vecs)
```

```python
# document embedding

docs = ([
    'Alpha is first letter in Greek',
    'Beta is second letter in Greek'
])

doc_vecs = embeds.embed_documents(docs)

len(doc_vecs[0]) , len(doc_vecs[1])
```
